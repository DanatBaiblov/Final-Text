##**Shifting Priorities: Evolution of Presidential Addresses in Kazakhstan**

**Background**

Kazakhstan is a presidential republic with a separation of powers among the executive, legislative, and judicial branches, according to the Constitution. Although the political system endorses a multi-party system, the ruling party's dominance has historically been a significant aspect of Kazakh politics. Since gaining independence in 1991, the constitutional framework has been established, laying the foundation for a political system where the President plays a central role in governing.

Within the executive branch of the Kazakhstani government, the President holds both the positions of head of state and head of government. This enables the President to exercise significant powers that include appointing and dismissing government officials and the Prime Minister. Additionally, the President plays a crucial role in shaping foreign policy, representing Kazakhstan on the global stage, and serving as the Commander-in-Chief of the Armed Forces. The President's dual role consolidates executive authority, resulting in a more effective functioning of the government.

The President has a range of important duties, such as designing foreign policy, leading diplomatic efforts, and ensuring national security as the Commander-in-Chief. Additionally, the President has the authority to appoint significant officials, including the Prime Minister, who can impact the composition and direction of the executive branch. Although the Parliament has the power to make laws, the President can suggest legislation and some decisions require the President's approval, adding to the executive's legislative function.

The political landscape in Kazakhstan has undergone evolution, notably during the era of Nursultan Nazarbayev (1991-2019), where he played a pivotal role in shaping the nation's political foundations. Despite fostering stability and economic growth, Nazarbayev's presidency faced scrutiny for limiting political pluralism.

The transition of power to Kassym-Jomart Tokayev in 2019, following Nursultan Nazarbayev's resignation, was initially perceived as a nominal shift due to several factors that constrained Tokayev's authority. Crucially, Nazarbayev retained a significant influence by assuming the role of the head of the Security Council, a position that held substantial decision-making powers, particularly in matters related to political appointments and security. This arrangement allowed Nazarbayev to wield considerable influence over key policy decisions and appointments, thereby mitigating the actual power transfer to Tokayev.

Compounding this, various sectors of the Kazakh economy were reportedly monopolized by members of Nazarbayev's family, further limiting Tokayev's ability to enact real and substantive reforms. The concentration of economic control within the family network impeded Tokayev's capacity to implement independent policy measures and address critical economic challenges facing the nation.

The nominal nature of the transition became evident in the absence of substantial reforms during the initial period of Tokayev's presidency. The political landscape remained largely unchanged, and the anticipated shifts in governance and economic policies did not materialize. This lack of tangible reform initiatives contributed to growing dissatisfaction among the Kazakh populace, leading to widespread discontent.

The culmination of these factors ultimately erupted in the January 2022 unrest. The protests were fueled by a variety of grievances, including economic challenges, political stagnation, and perceived inequalities. The public discontent underscored the need for genuine reforms that addressed these pressing issues, prompting a reevaluation of the existing power dynamics.

In response to the unrest, President Tokayev took decisive steps to consolidate real power. The removal of Nursultan Nazarbayev from his position as the head of the Security Council marked a significant turning point, stripping away a key source of influence over political and security matters. Additionally, the resignation of the government and the subsequent extension of a state of emergency provided Tokayev with the authority to restructure the political landscape.

The subsequent elections played a crucial role in solidifying Tokayev's mandate and legitimacy. With widespread support from the public, Tokayev secured a renewed term in office, providing him with the genuine political capital needed to implement meaningful reforms. The events surrounding the January unrest, therefore, served as a catalyst for Tokayev to obtain real power, enabling him to break free from the nominal transition and assert his authority in steering Kazakhstan towards a new era of governance and reform.


**Research Question**

How has the focus of reforms articulated in the annual addresses of the President of Kazakhstan to the people evolved over the past five years, and what patterns or shifts can be identified in the prioritization of key reform areas?

**Hypothesis**

The hypothesis posits that a significant shift in the focus of the last annual address of the second President of Kazakhstan has occurred, transitioning from a predominantly political emphasis to a pronounced emphasis on economic reforms.

This change is attributed to the altered political landscape in Kazakhstan, particularly the events surrounding the protests in January 2022. Prior to this period, the First President Nursultan Nazarbayev and his family wielded substantial influence, constraining the ability of the Second President Kassym-Jomart Tokayev to enact reforms in various economic sectors, which were reportedly monopolized by the First President's family. However, the unrest in January 2022, triggered by economic grievances and discontent with the government, culminated in significant political transformations, including the removal of Nursultan Nazarbayev from the chairmanship of the Security Council and the resignation of the previous government.

The hypothesis suggests that the ensuing political developments, including President Tokayev's reelection with popular support, have provided him with real power and autonomy to address economic challenges and conduct reforms. As a result, it is expected that the last annual address would reflect a pronounced shift towards economic considerations, as the President seizes the newfound authority to implement meaningful changes in the country's economic landscape.

```python
import requests
from bs4 import BeautifulSoup
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk import FreqDist
from sklearn.feature_extraction.text import TfidfVectorizer
import os
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data_df['URL']=data_df['URL'].fillna('no URL found') #checking if there are any null columns in the existing data.
```

Autamating the Retrieval of Data from URLs

```python
def scrape_article(url):
    response = requests.get(url)
    response.encoding = 'utf-8'
    html_string = response.text
    return html_string
```

```python
data_df['text'] = data_df['URL'].apply(scrape_article)
```

```python
data_df
```

Retrieving the text from each URL

```python
#Applying BeautifulSoup
for text in data_df['text']:
    soup = BeautifulSoup(text)
    speech = soup.get_text()
    print(speech)

def extract_text(html_text):
    soup = BeautifulSoup(html_text, 'html.parser')
    speech = soup.get_text()
    return speech

data_df['cleaned_text'] = data_df['text'].apply(extract_text)
```

```python

data_df
```

Data Cleaning

```python
data_df['word_count'] = data_df['cleaned_text'].apply(lambda x: len(str(x).split())) ##counting words
data_df.drop("word_count_cleaned", axis=1, inplace=True)

data_df
```

```python
# Function to filter and tokenize words, remove capitalization, numbers and punctuation
def filter_and_tokenize(text):
    text_tokens = [t.lower() for t in text.split() if t.isalpha()]
    return text_tokens

# Apply the function to each row of the 'text1' column
data_df['text1_tokens'] = data_df['cleaned_text'].apply(lambda x: filter_and_tokenize(x))

data_df
```

Removing Stopwords

```python
#creating a new column with the every speech without stopwords and counting the number of words
data_df['text1_stops'] = data_df['text1_tokens'].apply(lambda tokens: [t for t in tokens if t not in stops])
len(data_df["text1_stops"].iloc[0]), len(data_df["text1_stops"].iloc[1]), len(data_df["text1_stops"].iloc[2]),len(data_df["text1_stops"].iloc[3]), len(data_df["text1_stops"].iloc[4])
```

Lemmatizing Words

```python
wordnet_lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

data_df['text1_clean'] = data_df['text1_stops'].apply(lambda tokens: [wordnet_lemmatizer.lemmatize(t, get_wordnet_pos(t)) for t in tokens])

data_df['text1_clean']


```

```python
# Step 1: Convert the lists in 'text1_clean' to strings
data_df['text1_clean_str'] = data_df['text1_clean'].apply(lambda x: ' '.join(x))

# Step 2: Create a TfidfVectorizer
vectorizer = TfidfVectorizer()

# Step 3: Fit and transform the 'text1_clean_str' column to obtain the TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(data_df['text1_clean_str'])

# Step 4: Get feature names (terms) from the TfidfVectorizer
feature_names = vectorizer.get_feature_names_out()

# Step 5: Create a DataFrame to store the TF-IDF scores along with feature names
tfidf_data = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# Step 6: Display the most important terms for each document
for i in range(len(data_df)):
    # Get the TF-IDF scores for the i-th document
    tfidf_scores_i = tfidf_data.iloc[i]

    # Get the top 10 terms for the i-th document
    top_terms_i = tfidf_scores_i.sort_values(ascending=False).head(10)

    # Print the results
    print(f"\nTop terms for Document {i + 1}:")
    print(top_terms_i)
```

```python
from nltk.corpus import wordnet

def find_similar_words(word):
    similar_words = set()
    synsets = wordnet.synsets(word)
    for synset in synsets:
        similar_words.update([lemma.name() for lemma in synset.lemmas()])
    return similar_words

# Lists of economic and political terms
economic_terms = ['economic', "business", "economy", "industry", "financial", "investment", "energy", "production", "market"]
political_terms = ['president', "presidential",'political', 'election', "agency", 'institution', "policy", "election", "party", "electoral", "right"]

# Create new lists to store the extended terms
extended_economic_terms = economic_terms.copy()
extended_political_terms = political_terms.copy()

# Find similar words for each term and update the lists
for term in economic_terms:
    similar_words = find_similar_words(term)
    extended_economic_terms.extend(similar_words)

for term in political_terms:
    similar_words = find_similar_words(term)
    extended_political_terms.extend(similar_words)

# Apply the function to each row in the "text1_clean" column and print the results
for index, row in data_df.iterrows():
    for term in extended_economic_terms:
        similar_words = find_similar_words(term)
        print(f"Similar words for {term} in row {index + 1}: {similar_words}")

    for term in extended_political_terms:
        similar_words = find_similar_words(term)
        print(f"Similar words for {term} in row {index + 1}: {similar_words}")
```

```python
import matplotlib.pyplot as plt

# Function to count the number of occurrences of terms in a text
def count_term_occurrences(text, terms):
    return sum(1 for term in terms if term in text)

# Create lists to store the counts for each speech
economic_counts = []
political_counts = []

# Iterate through each row and count occurrences
for index, row in data_df.iterrows():
    text = row['text1_clean']

    # Count occurrences of economic terms
    economic_count = count_term_occurrences(text, extended_economic_terms)
    economic_counts.append(economic_count)

    # Count occurrences of political terms
    political_count = count_term_occurrences(text, extended_political_terms)
    political_counts.append(political_count)

# Plot the results for economic terms
plt.figure(figsize=(10, 5))
plt.bar(range(len(data_df)), economic_counts, label='Economic Terms', alpha=0.7, color='blue')
plt.xlabel('Speech Index')
plt.ylabel('Number of Economic Occurrences')
plt.title('Occurrences of Economic Terms in Each Speech')
plt.show()

# Plot the results for political terms
plt.figure(figsize=(10, 5))
plt.bar(range(len(data_df)), political_counts, label='Political Terms', alpha=0.7, color='red')
plt.xlabel('Speech Index')
plt.ylabel('Number of Political Occurrences')
plt.title('Occurrences of Political Terms in Each Speech')
plt.show()
```
